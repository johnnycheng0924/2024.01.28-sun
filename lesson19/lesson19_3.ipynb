```python
import numpy as np
import pandas as pd

scores = np.random.randint(1000,10000,size=(50,5))
df1 = pd.DataFrame(scores,
             columns=['國文', '英文', '數學', '地理', '歷史'],
             index=range(1,51))
df1.index.name='學號'
df1.columns.name="科目"
df1
```

```python
df1.style\
.format(thousands=',')
```

```python
df2 = df1.style \
.highlight_max(props='color:white; font-weight:bold; background-color:darkblue') \
.highlight_min(props='color:white; font-weight:bold; background-color:darkred')
df2.to_excel('3年2班.xlsx')
```
